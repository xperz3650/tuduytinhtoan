```python
def in_uoc_chung(a, b):
    l = min(a, b)
    print(f"Toàn bộ ước chung của {a} và {b} là:")
    for i in range(1, l + 1):
        if a % i == 0 and b % i == 0:
            print(i, end=" ")
    print() 
try:
    a = int(input("Nhập vào số nguyên dương a: "))
    b = int(input("Nhập vào số nguyên dương b: "))
    if a > 0 and b > 0:
        in_uoc_chung(a, b)
    else:
        print("Lỗi: Vui lòng nhập số lớn hơn 0.")
except ValueError:
    print("Lỗi: Giá trị nhập vào phải là một số nguyên.")
```

```python
def check(n):
    if n <= 0:
        return False
    while n % 2 == 0:
        n = n // 2
    return n == 1

try:
    n = int(input("Nhập vào một số nguyên dương n: "))
    if n > 0:
        if check(n):
            print(f"{n} LÀ số có dạng 2^k.")
        else:
            print(f"{n} KHÔNG phải là số có dạng 2^k.")
    else:
        print("Lỗi: Vui lòng nhập số nguyên dương (lớn hơn 0).")
except ValueError:
    print("Lỗi: Giá trị nhập vào phải là một số nguyên.")
```

```python
def cal(n):
    tong = 0
    while n > 0:
        tong += n % 10
        n //= 10      
    return tong

n = int(input("Nhập vào số nguyên dương n: "))
if n > 0:
    print(f"Tổng các chữ số của {n} là: {cal(n)}")
else:
    print("Vui lòng nhập một số nguyên dương!")
```
