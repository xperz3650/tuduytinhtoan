```python
def giai_phuong_trinh_bac_1(a, b):
    """
    Hàm giải phương trình bậc 1: ax + b = 0
    """
    if a == 0:
        if b == 0:
            return "Phương trình có vô số nghiệm."
        else:
            return "Phương trình vô nghiệm."
    else:
        x = -b / a
        return f"Phương trình có nghiệm duy nhất x = {x:.2f}"
print("--- Giải phương trình bậc 1: ax + b = 0 ---")
try:
    a = float(input("Nhập hệ số a: "))
    b = float(input("Nhập hệ số b: "))
    ket_qua = giai_phuong_trinh_bac_1(a, b)
    print("->", ket_qua)

except ValueError:
    print("Lỗi: Vui lòng nhập số thực hợp lệ cho a và b!")
```

```python
import math

def giai_phuong_trinh_bac_2(a, b, c):
    """
    Hàm giải phương trình bậc 2: ax^2 + bx + c = 0
    """
    if a == 0:
        if b == 0:
            if c == 0:
                return "Phương trình có vô số nghiệm."
            else:
                return "Phương trình vô nghiệm."
        else:
            x = -c / b
            return f"Phương trình trở thành bậc 1. Có nghiệm x = {x:.2f}"
    delta = b**2 - 4*a*c
    if delta < 0:
        return "Phương trình vô nghiệm (không có nghiệm thực)."
    elif delta == 0:
        x = -b / (2 * a)
        return f"Phương trình có nghiệm kép: x1 = x2 = {x:.2f}"
    else:
        x1 = (-b + math.sqrt(delta)) / (2 * a)
        x2 = (-b - math.sqrt(delta)) / (2 * a)
        return f"Phương trình có 2 nghiệm phân biệt:\nx1 = {x1:.2f}\nx2 = {x2:.2f}"
print("--- Giải phương trình bậc 2: ax^2 + bx + c = 0 ---")
try:
    a = float(input("Nhập hệ số a: "))
    b = float(input("Nhập hệ số b: "))
    c = float(input("Nhập hệ số c: "))
    ket_qua = giai_phuong_trinh_bac_2(a, b, c)
    print("->", ket_qua)
except ValueError:
    print("Lỗi: Vui lòng nhập số thực hợp lệ cho các hệ số!")
```

```python
def giai_he_phuong_trinh(a, b, m, c, d, n):
    D = a * d - b * c
    Dx = m * d - b * n
    Dy = a * n - m * c
    if D != 0:
        x = Dx / D
        y = Dy / D
        return f"Hệ có nghiệm duy nhất:\nx = {x:.2f}\ny = {y:.2f}"
    else:
        if Dx == 0 and Dy == 0:
            return "Hệ phương trình có vô số nghiệm."
        else:
            return "Hệ phương trình vô nghiệm."
print("--- Giải hệ phương trình: ax + by = m và cx + dy = n ---")
try:
    print("Nhập hệ số cho phương trình 1 (ax + by = m):")
    a = float(input("a = "))
    b = float(input("b = "))
    m = float(input("m = "))
    print("\nNhập hệ số cho phương trình 2 (cx + dy = n):")
    c = float(input("c = "))
    d = float(input("d = "))
    n = float(input("n = "))
    print("\nKết quả:")
    ket_qua = giai_he_phuong_trinh(a, b, m, c, d, n)
    print("->\n" + ket_qua)
except ValueError:
    print("\nLỗi: Vui lòng nhập số thực hợp lệ cho các hệ số!")
```

```python
def chuyen_doi_thoi_gian(tong_giay):
    gio = tong_giay // 3600
    phut = (tong_giay % 3600) // 60
    giay = tong_giay % 60
    return f"{gio} giờ, {phut} phút, {giay} giây"

print("--- Chuyển đổi thời gian từ giây sang giờ, phút, giây ---")
try:
    x = int(input("Nhập tổng số giây (x): "))
    if x < 0:
        print("-> Lỗi: Thời gian không thể là số âm. Vui lòng nhập số >= 0.")
    else:
        ket_qua = chuyen_doi_thoi_gian(x)
        print(f"-> {x} giây tương đương với: {ket_qua}")
except ValueError:
    print("-> Lỗi: Vui lòng nhập một số nguyên (không chứa chữ cái hay dấu thập phân)!")
```

```python
import math
def kiem_tra_vi_tri_diem(x_i, y_i, r, x_a, y_a):
    d = math.sqrt((x_a - x_i)**2 + (y_a - y_i)**2)
    if math.isclose(d, r, rel_tol=1e-9):
        return f"Khoảng cách d = {d:.2f}. Điểm A nằm TRÊN (thuộc) đường tròn."
    elif d < r:
        return f"Khoảng cách d = {d:.2f}. Điểm A nằm TRONG đường tròn."
    else:
        return f"Khoảng cách d = {d:.2f}. Điểm A nằm NGOÀI đường tròn."
print("--- Kiểm tra vị trí điểm A đối với đường tròn tâm I ---")
try:
    print("1. Nhập thông tin đường tròn:")
    x_i = float(input("   Tọa độ tâm x: "))
    y_i = float(input("   Tọa độ tâm y: "))
    r = float(input("   Bán kính R: "))
    if r < 0:
        print("\n-> Lỗi: Bán kính không thể là số âm!")
    else:
        print("\n2. Nhập thông tin điểm A:")
        x_a = float(input("   Tọa độ x: "))
        y_a = float(input("   Tọa độ y: "))
        ket_qua = kiem_tra_vi_tri_diem(x_i, y_i, r, x_a, y_a)
        print("\n-> Kết quả:", ket_qua)
except ValueError:
    print("\nLỗi: Vui lòng nhập số thực hợp lệ cho tọa độ và bán kính!")
```
