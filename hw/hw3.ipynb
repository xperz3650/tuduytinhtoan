```python
h = int(input("Nhập chiều cao h = "))
for i in range(1, h + 1):
    if i == 1:
        print(" " * (h - i) + "*")
    elif i == h:
        print("*" * (2 * h - 1))
    else:
        print(" " * (h - i) + "*" + " " * (2 * i - 3) + "*")
```

```python
a = int(input("Nhập số nguyên dương a: "))
la_so_chinh_phuong = False
i = 1
while i * i <= a:
    if i * i == a:
        la_so_chinh_phuong = True
        break 
    i += 1
if la_so_chinh_phuong:
    print(f"{a} là số chính phương.")
else:
    print(f"{a} KHÔNG phải là số chính phương.")
```

```python
n = int(input("Nhập số nguyên dương n: "))
S = 0
k = 0
while S + (k + 1) < n:
    k += 1
    S += k
print(f"Số k cần tìm là: {k}")
print(f"Kiểm tra: S({k}) = {S} < {n}")
```

```python
A = int(input("Nhập số nguyên dương A: "))
S = 0.0
n = 0
while S <= A:
    n += 1
    S += 1.0 / n

print(f"Số n nhỏ nhất cần tìm là: {n}")
print(f"Kiểm tra: Tổng lúc này là {S:.4f} > {A}")
```

```python
print("Nhập vào một dãy số nguyên (nhập -1 để dừng):")
max_val = None
min_val = None
while True:
    so = int(input("Nhập số: "))
    if so == -1:
        break  
    if max_val is None or min_val is None:
        max_val = so
        min_val = so
    else:
        if so > max_val:
            max_val = so
        if so < min_val:
            min_val = so
if max_val is not None:
    print(f"Số lớn nhất trong dãy vừa nhập: {max_val}")
    print(f"Số nhỏ nhất trong dãy vừa nhập: {min_val}")
else:
    print("Bạn chưa nhập số hợp lệ nào!")
```

```python
n = int(input("Nhập số nguyên dương n: "))
chan = 0
le = 0
t = n
while t > 0:
    chu_so = t % 10
    if chu_so % 2 == 0:
        chan += 1
    else:
        le += 1
    t //= 10
print(f"Số {n} có {chan} chữ số chẵn và {le} chữ số lẻ.")
```

```python
A = int(input("Nhập số nguyên dương A: "))
f1, f2 = 1, 1
while f2 <= A:
    f_tiep_theo = f1 + f2
    f1 = f2
    f2 = f_tiep_theo
print(f"Số Fibonacci lớn nhất không vượt quá {A} là: {f1}")
```
