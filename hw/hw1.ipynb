```python
c = input("Nhập vào một kí tự: ")
if len(c) == 1:
    ma_ascii = ord(c)
    if 65 <= ma_ascii <= 90:
        print("Kết quả:", chr(ma_ascii + 32))
    elif 97 <= ma_ascii <= 122:
        print("Kết quả:", chr(ma_ascii - 32))
    else:
        print("Kí tự không phải chữ cái thuộc bảng Latinh.")
```

```python

c = input("Nhập vào một chữ cái in hoa: ")
if len(c) == 1 and c.isupper():
    if c == 'A':
        print("Trường hợp đặc biệt: Chữ cái liền trước 'a' được quy ước là 'z'")
        print("Kết quả: z")
    else:
        c_lower = c.lower()
        ky_tu_truoc = chr(ord(c_lower) - 1)
        print(f"Ký tự viết thường tương ứng là '{c_lower}'")
        print(f"Kết quả (ký tự liền trước): {ky_tu_truoc}")
else:
    print("Vui lòng nhập đúng MỘT chữ cái in hoa (A-Z).")
```

```python
import math
a = float(input("Nhập số a: "))
b = float(input("Nhập số b: "))
c = float(input("Nhập số c: "))
if a > 0 and b > 0 and c > 0 and (a + b > c) and (a + c > b) and (b + c > a):
    p = (a + b + c) / 2
    dien_tich = math.sqrt(p * (p - a) * (p - b) * (p - c))
    print(f"{dien_tich:.1f}")
else:
    print("Khong phai 3 canh cua tam giac")
```
